```
# Dataset download: https://www.kaggle.com/sobhanmoosavi/us-accidents/download
import pandas as pd
import numpy as np
import plotly.graph_objects as go
import plotly.io as pio
from plotly.subplots import make_subplots
import warnings
warnings.filterwarnings('ignore')
```

```
US_Accidents_2016_to_2019 = pd.read_csv('US_Accidents_2016_to_2019.csv')
print(US_Accidents_2016_to_2019.shape)
US_Accidents_2016_to_2019.head()
```

```
US_Accidents_2016_to_2019.dtypes
```

```
def clean_zipcode(zipcode):
  if isinstance(zipcode, str):
    return int(zipcode[0:5])
  else:
    return 0
```

```
US_Accidents_2016_to_2019['Zipcode'] = US_Accidents_2016_to_2019['Zipcode'].apply(clean_zipcode)
```

```
US_Accidents_2016_to_2019.loc[US_Accidents_2016_to_2019['Visibility(mi)'] > 10.0, 'Visibility(mi)'] = 10.0
```

```
US_Accidents_2016_to_2019['Number'] = US_Accidents_2016_to_2019['Number'].fillna(1)
```

```
US_Accidents_2016_to_2019['Number'] = US_Accidents_2016_to_2019['Number'].astype('int64')
```

```
make_categorical = ['Severity', 'Side', 'Country', 'Timezone', 'Sunrise_Sunset', 'Civil_Twilight', 
 'Nautical_Twilight', 'Astronomical_Twilight']
```

```
def to_categorical(column_headers):
  for column in column_headers:
    US_Accidents_2016_to_2019[column] = US_Accidents_2016_to_2019[column].astype('category')
```

```
to_categorical(make_categorical)
```

```
US_Accidents_2016_to_2019['Start_Time'] = pd.to_datetime(US_Accidents_2016_to_2019['Start_Time'], infer_datetime_format=True)
US_Accidents_2016_to_2019['End_Time'] = pd.to_datetime(US_Accidents_2016_to_2019['End_Time'], infer_datetime_format=True)
```

```
US_Accidents_2016_to_2019['Year'] = US_Accidents_2016_to_2019['Start_Time'].dt.year
US_Accidents_2016_to_2019['Month'] = US_Accidents_2016_to_2019['Start_Time'].dt.month
```

```
month_dictionary = {1: 'January', 2: 'February', 3: 'March', 4: 'April', 
                    5: 'May', 6: 'June', 7: 'July', 8: 'August', 9: 'September', 
                    10: 'October', 11: 'November', 12: 'December'}
```

```
US_Accidents_2016_to_2019 = US_Accidents_2016_to_2019.replace({'Month': month_dictionary})
```

```
US_Accidents_2016_to_2019['Month'] = US_Accidents_2016_to_2019['Month'].astype('category')
```

```
US_Accidents_2018 = US_Accidents_2016_to_2019[US_Accidents_2016_to_2019['Year'] == 2018]
US_Accidents_2018.reset_index(drop=True, inplace=True)
```

```
US_Accidents_2018.head()
```

```
US_Accidents_2018.dtypes
```

```
states = US_Accidents_2018.State.unique()
```

```
accidents_by_state = []
for state in states:
  truth = US_Accidents_2018['State'] == state
  count = US_Accidents_2018[truth].count()
  accidents_by_state.append(count['ID'])
```

```
temperature = US_Accidents_2018['Temperature(F)']
severity_1 = US_Accidents_2018[US_Accidents_2018['Severity']==1]['Temperature(F)'].mean()
severity_2 = US_Accidents_2018[US_Accidents_2018['Severity']==2]['Temperature(F)'].mean()
severity_3 = US_Accidents_2018[US_Accidents_2018['Severity']==3]['Temperature(F)'].mean()
severity_4 = US_Accidents_2018[US_Accidents_2018['Severity']==4]['Temperature(F)'].mean()
severity_labels = ['Severity 1', 'Severity 2', 'Severity 3', 'Severity 4']

mean_temp_by_severity = [severity_1, severity_2, severity_3, severity_4]
```

```
weather = US_Accidents_2018.Weather_Condition.value_counts()
```

```
population = pd.read_excel('Population.xlsx')
keys = population['State'].values.tolist()
values = population['Population'].values.tolist()
dictionary = dict(zip(keys, values))
US_Accidents_2018['Population'] = US_Accidents_2018['State'].map(dictionary)
```

```
per_capita_state_accidents = accidents_by_state/US_Accidents_2018['Population'].unique()
```

```
October = US_Accidents_2018[US_Accidents_2018['Month'] == 'October']
```

```
year = [2018]
distance_traffic = [US_Accidents_2018['Distance(mi)'].sum()]
distance_moon = [238900]
```

```
pio.templates.default = 'plotly_dark'
fig1 = go.Figure()
fig1.add_trace(go.Bar(x=severity_labels,
                y= mean_temp_by_severity,
                name='Distance in Traffic',
                marker_color='#BFFF00',
                width = 0.2
                ))
fig1.update_layout(
    title='<b>Accident Severity vs Mean Temperature</b>', 
    font=dict(
        size=14,
        color='#F2F2F2'),
    yaxis=dict(
        title='Temperature (F)',
        titlefont_size=16,
        tickfont_size=14,
        color='#F2F2F2'
              ),
    xaxis=dict(
        title='Severity',
        titlefont_size=16,
        tickfont_size=14,
        color='#F2F2F2'
              ),
    legend=dict(
        x=1,
        y=1,
        bgcolor='rgba(255, 255, 255, 0)',
        bordercolor='rgba(255, 255, 255, 0)'
    ), 
    barmode='group',
    bargap=0.4,
    bargroupgap=0.15
)
fig1.show()
```

```
fig2 = go.Figure()
fig2.add_trace(go.Bar(x=weather.values,
                y= weather.index,
                name='Distance in Traffic',
                orientation='h',
                marker_color='#BFFF00',
                width = 0.5
                ))
fig2.update_layout(
    title="<b>Weather Conditions as Related to Accidents</b>", 
    font=dict(
        size=14,
        color='#F2F2F2'),
    yaxis=dict(
        title='Weather Condition',
        titlefont_size=16,
        tickfont_size=12,
        color='#F2F2F2'
              ),
    xaxis=dict(
        title='Accidents',
        titlefont_size=16,
        tickfont_size=14,
        color='#F2F2F2'
              ),
    legend=dict(
        x=1,
        y=1,
        bgcolor='rgba(255, 255, 255, 0)',
        bordercolor='rgba(255, 255, 255, 0)'
    ), 
    barmode='group',
    bargap=0.4,
    bargroupgap=0.15
)
fig2.show()
```

```
fig3 = go.Figure(data=go.Choropleth(
    locations=US_Accidents_2018['State'].unique(),
    z = accidents_by_state, 
    locationmode = 'USA-states', 
    colorscale = 'Inferno',
    colorbar_title = 'Number of Accidents',
))

fig3.update_layout(
    title_text = '<b>2018 US Accidents by State</b>', 
    geo_scope='usa', font=dict(
        size=14,
        color="#F2F2F2",
))

fig3.show()
```

```
fig4 = make_subplots(specs=[[{'secondary_y': True}]])

fig4.add_trace(go.Bar(x=states,
                y= accidents_by_state,
                name='Accidents',
                marker_color='#0085A1'
                ))
fig4.add_trace(
    go.Scatter(x=states, y= US_Accidents_2018['Population'].unique(), 
               mode= 'lines+markers', name='Population', 
               line=dict(color='crimson', width=4)),
    secondary_y=True,
)

fig4.update_layout(
    title_text='<b>2018 US Accidents by State and Population</b>', font=dict(
        size=14,
        color='#F2F2F2', 
))

fig4.update_xaxes(title_text='State', color='#F2F2F2')

fig4.update_yaxes(title_text='Accidents', color='#F2F2F2', secondary_y=False)
fig4.update_yaxes(title_text='Population', color='#F2F2F2',secondary_y=True)

fig4.show()
```

```
fig6 = go.Figure(data=go.Scattergeo(
        locationmode = 'USA-states',
        lon = October['Start_Lng'],
        lat = October['Start_Lat'],
        mode = 'markers',
        marker = dict(
            size = 8,
            opacity = 0.8,
            reversescale = True,
            autocolorscale = False,
            symbol = 'square',
            line = dict(
                width=1,
                color='rgba(102, 102, 102)'
            ),
            colorscale = 'Viridis',
            cmin = 0,
            color = October['Population'],
            cmax = October['Population'].max(),
            colorbar_title='State Population<br>October 2018'
        )))

fig6.update_layout(
        title = '<b>October 2018 US Accident Coordinates<br>87,192 of 892,615 recorded 2018 Accidents<br>(Heaviest Month by Annual Total)</b>', 
        font=dict(
        size=14,
        color='#F2F2F2'),
        geo = dict(
            scope='usa',
            projection_type='albers usa',
            showland = True,
            landcolor = "#F2F2F2",
            subunitcolor = "rgb(217, 217, 217)",
            countrycolor = "rgb(217, 217, 217)",
            countrywidth = 0.5,
            subunitwidth = 0.5
        ),
    )
fig6.show()
```
